<big>**基本介绍**</big>

Logistic 回归的本质是：**假设数据服从这个分布**，然后使用极大似然估计做参数的估计。

<big>**线性回归**</big>


线性回归是一个回归问题，要预测的变量 y 与自变量 x 的关系是线性的。

比如，给定N对数据(x, y)，寻找最优参数w和b，使得函数可以更好地拟合数据。





损失函数采用预测值与实际值的误差的平方，即$L= \frac{1}{N} \displaystyle \sum_{i=1}^{N}(x_i*w+b-y_i)^2$



上图中，线段表示的是预测值与实际值的误差，平方后，也就是正方形。所有正方形的面积和最小时，就是损失最小。


最小二乘法中，二乘就是平方，最小就是损失函数最小。使用平方的原因在于平方函数有最小值，能用求导的方式得到最小值，绝对值函数则不保证有这种属性。

在得到损失函数后，为了得到最小值，需要对参数进行求导。



通过令偏导为0得到最小值。从而求出最优参数。


在多元线性回归模型经典假设中，其重要假定之一是回归模型的解释变量之间不存在线性关系。当有多维特征时，则首先需要判断是否存在多重共线性。

多重共线性问题就是指一个解释变量的变化引起另一个解释变量地变化，即某特征是其他部分特征的线性组合。

判断多重共线性的指标：VIF（严格大于5即认为存在严重的多重共线性）、容差、相关系数和显著性等。

当存在共线性特征时，可以采用

1、手动剔除

2、岭回归
